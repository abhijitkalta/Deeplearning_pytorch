```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python
n1 = 30
n2 = 40
mu1 = 1
mu2 = 2

data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

plt.plot(np.zeros(n1), data1, 'ro', markerfacecolor = 'w')
plt.plot(np.ones(n2), data2, 'bs', markerfacecolor = 'w')
plt.rcParams.update({'font.size': 18})
plt.xticks([0,1], labels = ['Group1', 'Group2']);     
         
```

```python
t, p = stats.ttest_ind(data1, data2)
print(t, p)
```

```python
# Derivative 
import sympy as sym
import sympy.plotting.plot as sympyplot
```

```python
x = sym.symbols('x')
fx = 2*x**2
df = sym.diff(fx, x)
print(fx)
print(df)
```

```python
sympyplot(fx, (x, -4, 4), title = 'the function');
```

```python
sympyplot(df, (x, -4, 4), tile = 'the derivative');
```
